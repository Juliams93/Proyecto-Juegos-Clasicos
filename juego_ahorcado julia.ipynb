# Juego del Ahorcado

```python
import random
# Lista de palabras para adivinar
palabras = ["aguacate", "python", "ahorcado"]
```

```python
def iniciar_juego():
    partida = input('¿Quieres jugar?: Sí (S) No(N)').lower()
    if partida == 's':
        print('¡¡Bienvenido al juego del ahorcado!!!')
    else: 
        print('Hasta pronto')
```

```python
def mostrar_ahorcado(intentos):
    if intentos == 0:
        print('''
           ------
           |    |
           |
           |
           |
           |
        --------
        ''')
    elif intentos == 1:
        print('''
           ------
           |    |
           |    O
           |
           |
           |
        --------
        ''')
    elif intentos == 2:
        print('''
           ------
           |    |
           |    O
           |    |
           |
           |
        --------
        ''')
    elif intentos == 3:
        print('''
           ------
           |    |
           |    O
           |   /|
           |
           |
        --------
        ''')
    elif intentos == 4:
        print('''
           ------
           |    |
           |    O
           |   /|\\
           |
           |
        --------
        ''')
    elif intentos == 5:
        print('''
           ------
           |    |
           |    O
           |   /|\\
           |   /
           |
        --------
        ''')
    elif intentos == 6:
        print('''
           ------
           |    |
           |    O
           |   /|\\
           |   / \\
           |
        --------
        ''')

```

```python
# Función para obtener las posiciones de la letra en la palabra
def adivinar_posiciones(palabra, letra):
    posiciones = [i for i, char in enumerate(palabra) if char == letra]
    return posiciones
```

```python
def validar_letras(letra, letras_adivinadas):
    try:
        if len(letra) != 1 or not letra.isalpha(): # Si la longitud de la letra es diferente a 1 o no es una letra.
            print('Por favor introduce una sola letra válida')
            return False  # Sale de la función y devuelve False si la letra no es válida
        elif letra in letras_adivinadas:
            print('Ya has adivinado esta letra. Intenta con otra')
            return False # Sale de la función y devuelve False si la letra ya fue usada
        return True # Si pasa ambas verificaciones, la letra es válida # Sale de la función y devuelve True si la letra es válida
    except TypeError:
        print('Introduce una letra, por favor')
        return False # Sale de la función y devuelve False si ocurre un error de tipo
```

```python

def desarrollo_juego():
    max_intentos = 6  # Definimos el máximo de intentos
    intentos_fallidos = 0
    palabra = random.choice(palabras)
    palabra_oculta = "_" * len(palabra)  # Palabra oculta con guiones bajos
    letras_adivinadas = []  # Lista para letras ya adivinadas
    while intentos_fallidos < max_intentos:
        print(f'Palabra oculta: {palabra_oculta}')
        mostrar_ahorcado(intentos_fallidos)
        letra = input('Escribe una letra: ').lower()
         
        # Verificamos si la letra es válida antes de continuar
        if not validar_letras(letra, letras_adivinadas):
            intentos_fallidos += 1
            print(f'Intentos fallidos: {intentos_fallidos}/{max_intentos}')
            continue  # Si no es válida, volvemos a pedir la letra
        
        letras_adivinadas.append(letra)
        # Usamos la función adivinar_posiciones para encontrar la letra en la palabra
        posiciones = adivinar_posiciones(palabra, letra)
        if posiciones:
            print(f'¡La letra "{letra}" está en la palabra!')
            # Actualizamos la palabra oculta con las posiciones encontradas
            palabra_oculta_lista = list(palabra_oculta)  # Convertimos a lista para modificar
            for posicion in posiciones:
                palabra_oculta_lista[posicion] = letra  # Reemplazamos los guiones bajos con la letra
            palabra_oculta = "".join(palabra_oculta_lista)  # Convertimos de nuevo a cadena
            print(f'Palabra oculta actualizada: {palabra_oculta}')
            # Si ya no quedan guiones bajos, el jugador ha ganado
            if "_" not in palabra_oculta:
                print(f'¡Felicidades! Has adivinado la palabra: {palabra}')
                break
        else:
            # Si no adivina, incrementamos los intentos fallidos
            print(f'La letra "{letra}" no está en la palabra. Prueba de nuevo.')
            intentos_fallidos += 1
            print(f'Intentos fallidos: {intentos_fallidos}/{max_intentos}')
    # Condición si se acaban los intentos
    if intentos_fallidos == max_intentos:
        print(f'Máximo de intentos agotados.')
    return palabra_oculta, palabra, max_intentos 
```

```python
def jugar ():
    iniciar_juego()
    palabra_oculta, palabra, max_intentos = desarrollo_juego()
    terminar_juego(palabra_oculta, palabra, max_intentos)
```

```python
def jugar_otra_vez():
   # Bucle principal para jugar otra vez
    while True:
        jugar()  # Inicia una partida
        jugar_otra_vez = input("¿Quieres jugar otra vez? (S/N): ").lower()
        if jugar_otra_vez != 's':
            print("¡Gracias por jugar!")
            break
```
